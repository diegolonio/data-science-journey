<h2>ÁLGEBRA LINEAL</h2>

Si $AB = I = BA$ se dice que $B$ es la inversa de $A$ y se denota como $A^1$

```python
A = np.array([
    [1, 2, 1],
    [4, 4, 5],
    [6, 7, 7]
])

B = np.array([
    [-7, -7, 6],
    [2, 1, -1],
    [4, 5, -4]
])
```

```python
# Multiplicamos las matrices de ambas formas
AB = np.dot(A, B)
BA = np.dot(B, A)
```

```python
print(AB)
```

```python
print(BA)
```

```python
print(np.array_equal(AB, BA))
```

La transpuesta de una matriz se obtiene cuando conviertes sus filas en columnas y se escribe como $A^T$

```python
print(A)
```

```python
print(A.T)
```

Algunas propiedades de las inversas y transpuestas son:

```python
A = np.array([[1, 2, 0],
              [0, 1, 3],
              [4, 0, 1]])

B = np.array([[2, 0, 1],
              [1, 1, 0],
              [3, 1, 2]])
```

$AA^{-1} = I = A^{-1}A$ (la verificamos hace un momento)

$ (AB)^{-1} = B^{-1}A^{-1} $

```python
print(CL := np.linalg.inv(A @ B))
```

```python
print(CR := np.linalg.inv(B) @ np.linalg.inv(A))
```

```python
print(np.allclose(CL, CR))
```

$ (A + B)^{-1} \neq A^{-1} + B^{-1} $

```python
print(CL := np.linalg.inv(A + B))
```

```python
print(CR := np.linalg.inv(A) + np.linalg.inv(B))
```

```python
print(np.allclose(CL, CR))
```

$ (A^T)^T = A $

```python
print(A)
```

```python
print(A_T := A.T)
```

```python
print(A_T.T)
```

```python
print(np.array_equal(A, A_T.T))
```

$ (AB)^T = B^TA^T $

```python
print(CL := (A @ B).T)
```

```python
print(CR := B.T @ A.T)
```

```python
print(np.array_equal(CL, CR))
```

$ (A + B)^T = A^T + B^T $

```python
print(CL := (A + B).T)
```

```python
print(CR := A.T + B.T)
```

```python
print(np.array_equal(CL, CR))
```

Una matriz simétrica es la que $A = A^T$, nótese que esto solo puede pasar con matrices cuadradas. La suma de dos matrices simétricas es también simétrica, el producto no siempre.
$$\begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix}
\begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix}
=
\begin{bmatrix} 1 & 1 \\ 0 & 0 \end{bmatrix}$$
Por otro lado, si $A$ es invertible entonces $A^T$ también lo es. De modo que $(A^T)^{-1} = (A^{-1})^T$

```python
print(CL := np.linalg.inv(A.T))
```

```python
print(CR := (np.linalg.inv(A)).T)
```

```python
print(np.allclose(CL, CR))
```
