```python
arr = np.zeros((8, 4))
arr
```

```python
for i in range(8):
    arr[i] = str(i)

arr
```

```python
arr[[-1, -7]]
```

```python
arr = np.arange(32).reshape((8, 4))
arr
```

```python
arr[[1, 5, 7, 2], [0, 3, 1, 2]]
```

```python
arr2 = arr[[1, 5, 7, 2]][:, [0, 3, 1, 2]]
arr2
```

```python
arr[[1, 5, 7, 2]][:, [0, 3, 1, 2]] = 100
arr
```
